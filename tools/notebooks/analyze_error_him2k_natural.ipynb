```python
import os
import glob
import numpy as np
import cv2
```

```python
alpha_dir = '/home/chuongh/vm2m/data/HIM2K/alphas/natural'
baseline_dirs = [
    '/home/chuongh/vm2m/data/HIM2K/instmatt_public_r50_fpn_3x/natural',
    '/home/chuongh/vm2m/data/HIM2K/instmatt_r50_fpn_3x/natural',
    '/home/chuongh/vm2m/output/HIM/baselines/mgm_single-mask/mgm_cvpr23_multi-inst_him_bs12_0920/vis_r50_fpn_3x/natural',
    '/home/chuongh/vm2m/output/HIM/baselines/mgm_stacked-mask/mgm_cvpr23_stacked_multi-inst_him_bs12_0920/vis_r50_fpn_3x/natural',
    '/home/chuongh/sensei-fs-symlink/users/chuongh/vm2m/output/HIM/ours_1108_stronger-aug_guidance_scratch/vis_24.5k_r50_fpn_3x/natural/natural'
]
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (30, 30))
f_mad = open('mad_him2k_natural_1110.csv', 'w')
f_mad_bg = open('mad_bg_him2k_natural_1110.csv', 'w')
f_mad_fg = open('mad_fg_him2k_natural_1110.csv', 'w')
f_mad_unk = open('mad_unk_him2k_natural_1110.csv', 'w')
for image_name in os.listdir(alpha_dir):
    print(image_name)
    alpha_names = glob.glob(os.path.join(alpha_dir, image_name) + '/*.png')

    mad = []
    mad_bg = []
    mad_fg = []
    mad_unk = []
    for alpha_name in alpha_names:
        alpha_gt = cv2.imread(alpha_name, cv2.IMREAD_GRAYSCALE) / 255.0
        trimap = np.zeros_like(alpha_gt)
        trimap[alpha_gt >= 254.0/255.0] = 2.0
        unknown_region = np.logical_and(alpha_gt > 1.0/255.0, alpha_gt < 254.0/255.0)
        unknown_region = cv2.dilate(unknown_region.astype('uint8'), kernel)
        trimap[unknown_region > 0] = 1.0

        for baseline in baseline_dirs:
            alpha_pred = cv2.imread(alpha_name.replace(alpha_dir, baseline), cv2.IMREAD_GRAYSCALE) / 255.0

            diff = np.abs(alpha_gt - alpha_pred)
            mad.append(np.mean(diff))

            mad_bg.append(np.mean(diff[trimap == 0]))
            mad_fg.append(np.mean(diff[trimap == 2.0]))
            mad_unk.append(np.mean(diff[trimap == 1.0]))
    mad = np.array(mad).reshape(-1, len(baseline_dirs))
    mad_bg = np.array(mad_bg).reshape(-1, len(baseline_dirs))
    mad_fg = np.array(mad_fg).reshape(-1, len(baseline_dirs))
    mad_unk = np.array(mad_unk).reshape(-1, len(baseline_dirs))
    mad = mad.mean(0)
    mad_bg = mad_bg.mean(0)
    mad_fg = mad_fg.mean(0)
    mad_unk = mad_unk.mean(0)
    f_mad.write(image_name + ',' + ','.join([str(x) for x in mad]) + '\n')
    f_mad_bg.write(image_name + ',' + ','.join([str(x) for x in mad_bg]) + '\n')
    f_mad_fg.write(image_name + ',' + ','.join([str(x) for x in mad_fg]) + '\n')
    f_mad_unk.write(image_name + ',' + ','.join([str(x) for x in mad_unk]) + '\n')

f_mad.close()
f_mad_bg.close()
f_mad_fg.close()
f_mad_unk.close()
```
