```python
import os
import glob
```

```python
pha_dir = "/home/chuongh/vm2m/data/HIM2K/alphas"
mask_dir = "/home/chuongh/vm2m/data/HIM2K"
```

```python
import random
import cv2
import numpy as np

# function to process each image
def process_image(src_path, dst_path):
    os.makedirs(os.path.dirname(dst_path), exist_ok=True)


    # read the image in grayscale format
    img = cv2.imread(src_path, cv2.IMREAD_GRAYSCALE)

    # normalize image to 0-1
    img_normalized = img / 255.0

    # generate a random threshold to binarize the image, excluding 0 and 1
    threshold = random.uniform(0.01, 0.99)

    # binarize the image
    _, img_binarized = cv2.threshold(img_normalized, threshold, 1, cv2.THRESH_BINARY)

    # generate random kernel sizes for dilation and erosion
    kernel_size_dilate = random.randint(1, 30)
    kernel_size_erode = random.randint(1, 30)

    # create the kernels
    kernel_dilate = np.ones((kernel_size_dilate, kernel_size_dilate), np.uint8)
    kernel_erode = np.ones((kernel_size_erode, kernel_size_erode), np.uint8)

    # randomly decide the order of dilation and erosion and whether to do both or just one
    operation_order = random.choice(["dilate_erode", "erode_dilate", "dilate", "erode"])

    if operation_order == "dilate_erode":
        img_dilated = cv2.dilate(img_binarized, kernel_dilate, iterations=1)
        img_final = cv2.erode(img_dilated, kernel_erode, iterations=1)
    elif operation_order == "erode_dilate":
        img_eroded = cv2.erode(img_binarized, kernel_erode, iterations=1)
        img_final = cv2.dilate(img_eroded, kernel_dilate, iterations=1)
    elif operation_order == "dilate":
        img_final = cv2.dilate(img_binarized, kernel_dilate, iterations=1)
    else:  # operation_order == "erode"
        img_final = cv2.erode(img_binarized, kernel_erode, iterations=1)

    # save the image in the destination directory
    cv2.imwrite(dst_path, img_final * 255)
```

```python

all_pha_names = glob.glob(pha_dir + "/*/*/*.png")
for mask_dir in glob.glob("/home/chuongh/vm2m/data/HIM2K/masks_matched_*"):
    for pha_path in all_pha_names:
        mask_path = pha_path.replace(pha_dir, mask_dir)
        if os.path.exists(mask_path):
            continue
        
        # Create mask by random perturb the alpha
        print("Generate",mask_path)
        process_image(pha_path, mask_path)
```
