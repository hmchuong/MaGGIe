```python
import os
import shutil
import glob
```

```python
source_dir = "/home/chuongh/vm2m/data/VHM/benchmark/real_qual"
target_dir = "/home/chuongh/vm2m/data/VHM/benchmark/real_qual_filtered"
```

```python
for video_name in os.listdir(source_dir + "/fgr"):
    video_dir = os.path.join(source_dir, "fgr", video_name)
    frame_names = glob.glob(video_dir + "/*.jpg")
    frame_names.sort()
    frame_names = frame_names[::5]
    mask_dir = os.path.join(source_dir, "mask", video_name)
    for frame_name in frame_names:
        tgt_path = frame_name.replace(source_dir, target_dir)
        os.makedirs(os.path.dirname(tgt_path), exist_ok=True)
        shutil.copy2(frame_name, tgt_path)
        mask_dir = frame_name.replace("/fgr/", "/mask/").replace(".jpg", "")
        tgt_mask_dir = mask_dir.replace(source_dir, target_dir)
        # os.makedirs(tgt_mask_dir, exist_ok=True)
        shutil.copytree(mask_dir, tgt_mask_dir)
```

```python
# rename some mask to continuous

mask_dir = '/home/chuongh/vm2m/data/VHM/benchmark/real_qual_filtered/mask'

for video_name in os.listdir(mask_dir):

    mask_names = glob.glob(os.path.join(mask_dir, video_name) + "/*/*.png")
    unique_mask_names = set([os.path.basename(mask_name) for mask_name in mask_names])
    unique_mask_names = sorted(list(unique_mask_names))
    new_names = {name: "{:02d}.png".format(i) for i, name in enumerate(unique_mask_names)}

    for old_name, new_name in new_names.items():
        if old_name != new_name:
            for mask_name in mask_names:
                if old_name in mask_name:
                    os.rename(mask_name, mask_name.replace(old_name, new_name))
                    print(mask_name, mask_name.replace(old_name, new_name))
```
